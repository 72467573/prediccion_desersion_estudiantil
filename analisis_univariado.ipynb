## Examinando una variable a la vez:

```python
# Cargar nuevamente los datos y definir tipos de variables
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
file_path = "dataset.csv"
df = pd.read_csv(file_path)

# Identificar variables numéricas y categóricas
numeric_vars = df.select_dtypes(include=['number']).columns
categorical_vars = df.select_dtypes(exclude=['number']).columns

# Crear estructuras para almacenar resúmenes
numeric_summaries = {}
categorical_summaries = {}

# Crear gráficos y resúmenes para cada variable
for col in df.columns:
    plt.figure(figsize=(10, 4))
    if col in numeric_vars:
        # Resumen estadístico
        mean = df[col].mean()
        median = df[col].median()
        std = df[col].std()
        numeric_summaries[col] = {"mean": mean, "median": median, "std": std}

        # Gráficos
        plt.subplot(1, 2, 1)
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f'Histograma de {col}')

        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot de {col}')

    elif col in categorical_vars:
        # Frecuencia
        freq = df[col].value_counts()
        categorical_summaries[col] = freq

        # Gráfico de barras
        freq.plot(kind='bar')
        plt.title(f'Frecuencia de {col}')
        plt.ylabel("Frecuencia")
        plt.xlabel(col)
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

numeric_summaries, categorical_summaries
```

### Variables numericas:

```python
print("Variables numéricas:")
for var in numeric_vars:
    print(f"- {var}")
```

## Variables categóricas:

```python
print("\nVariables categóricas:")
for var in categorical_vars:
    print(f"- {var}")
```
